*Testing for Normality on Target variable*

```python
target_var = df['concrete_compressive_strength']
target_var.hist()
```

*skewed towards left*

```python
from scipy.stats.mstats import normaltest
```

```python
normaltest(target_var)
```

*pvalue is too low, this is not a normal distribution* 

*testing log, sqrt, boxcox transforms*

```python
target_var_log = np.log(target_var)
```

```python
plt.hist(target_var_log)
```

```python
target_var_sqrt = np.sqrt(target_var)
```

```python
plt.hist(target_var_sqrt)
```

```python
normaltest(target_var_sqrt)
```

```python
from scipy.stats import boxcox
```

```python
target_var_bc = boxcox(target_var)
```

```python
normaltest(target_var_bc[0])
```

```python
plt.hist(target_var_bc[0])
```

*conclusion: sqrt gives the most normal distribution*
