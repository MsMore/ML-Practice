```python
import os
data_path = ['data']
```

```python
import pandas as pd
import numpy as np

filepath = os.sep.join(data_path + ['winequality-red.csv'])
data = pd.read_csv(filepath, sep = ';')
```

*comments: no null value*

```python
df = data.rename(columns = {'fixed acidity': 'fixed_acidity', 'volatile acidity': 'volatile_acidity', 'citric acid': 'citric_acid', 'residual sugar': 'residual_sugar', 'free sulfur dioxide': 'free_sulfur_dioxide', 'total sulfur dioxide': 'total_sulfur_dioxide'})
```

*comments: volatile acidity decreases with quality. citric acid increases with quality. chlorides decrease with quality. Sulphates increase with quality.*


```python
import seaborn as sns
sns.set()
```

```python
sns.pairplot(df, height=1.2, aspect=1.25);
```

```python
from scipy.stats.mstats import normaltest
def isNormal(a):
    return normaltest(a)[1] >= 0.05
isNormal(df['quality'])
df['quality'].hist()
```

```python
isNormal(df['quality'].apply(np.log1p))
```

```python
df['quality'].apply(np.log1p).hist()
```

```python
isNormal(df['quality'].apply(np.sqrt))
```

```python
df['quality'].apply(np.sqrt).hist()
```

```python
from scipy.stats import boxcox
```

```python
isNormal(boxcox(df['quality'])[0])
```
